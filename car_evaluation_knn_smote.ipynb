```python
from sklearn.preprocessing import LabelEncoder
```

```python
features_tb_encoded = car_ev.columns.difference(['doors', 'persons'])
features_tb_encoded
```

```python
car_ev[features_tb_encoded]
```

```python
label_enc = LabelEncoder()
car_ev_encoded = car_ev[features_tb_encoded].apply(label_enc.fit_transform)
```

```python
car_ev_encoded = pd.concat([car_ev_encoded, car_ev[['doors', 'persons']]], axis= 1)
car_ev_encoded
```

```python
car_ev_encoded['outcome'].value_counts()
```

```python
X = car_ev_encoded[car_ev_encoded.columns.difference(['outcome'])]
X
```

```python
Y = car_ev_encoded['outcome']
Y
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE()
X_res, Y_res = sm.fit_resample(X,Y)
```

```python
Y_res.value_counts()
```

```python
X_res
```

```python
Y_res
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_res, Y_res, test_size = 0.40, random_state= 43)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
train_preds = knn.predict(x_train)
train_preds
```

```python
test_preds = knn.predict(x_test)
test_preds
```

```python
knn.score(x_test, y_test)
```

```python
knn.score(x_train, y_train)
```
